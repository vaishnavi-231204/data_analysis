```python
data.fillna(0,inplace=True)
data
```

```python
def assign_grade(total):
    if total >= 60:
        return 'A'
    elif total >=50:
        return 'B'
    elif total >= 30:
        return 'C'
    elif total >= 20:
        return 'D'
    else:
        return 'F'

data['Grade'] = data['Total'].apply(assign_grade)



updated_file_path_with_grades = 'class_marks.csv'
data.to_csv(updated_file_path_with_grades, index=True)

print(f"Updated dataset with grades saved successfully to: {updated_file_path_with_grades}")
```
